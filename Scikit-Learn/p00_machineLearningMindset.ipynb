### What is Machine Learning?

Machine Learning = finding patterns from data to make predictions.

- You don’t write rules

- You show examples

- The model learns a pattern

For Example:

- You don’t code: “if area > 1000 and rooms > 3 then price = …”

- You give past house data → model learns the relationship

### Supervised vs Unsupervised

| Type             | Data Has Output? | Example                       |
| ---------------- | ---------------- | ----------------------------- |
| **Supervised**   | ✅ Yes            | House price, spam detection   |
| **Unsupervised** | ❌ No             | Customer grouping, clustering |


### Features vs Target

| Term             | Meaning       |
| ---------------- | ------------- |
| **Features (X)** | Input columns |
| **Target (y)**   | Output column |


Example (House prices):

| Area | Rooms | Price |
| ---- | ----- | ----- |
| 1200 | 3     | 50L   |

X = [Area, Rooms] <br>
y = Price

### What does a “model” actually do?

A model:

- Sees patterns

- Learns parameters

- Minimizes error

It does NOT:

- Think

- Understand meaning

- Know logic

It only knows numbers & relationships

### Why we split data (train / test)?

| Split             | Purpose                   |
| ----------------- | ------------------------- |
| **Training data** | Learn patterns            |
| **Testing data**  | Check if learning is real |


### Overfitting vs Underfitting

| Term             | Meaning                    |
| ---------------- | -------------------------- |
| **Underfitting** | Too simple, learns nothing |
| **Overfitting**  | Too complex, memorizes     |

Goal: generalization <br>
(Works on unseen data)

### Work Flow

1. Load data
2. Split into X and y
3. Train-test split
4. Choose model
5. fit()
6. predict()
7. evaluate()
