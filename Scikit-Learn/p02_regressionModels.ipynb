## Underfitting vs Overfitting

### Underfitting

Model is too simple to capture the pattern.

- Learns almost nothing

- High error on train and test

### Overfitting

Model is too complex and memorizes data.

- Very low error on training
- Very high error on test
- Fails on new data

### Good Fit

Captures the true pattern, not noise.

- Balanced complexity

- Low train error

- Low test error

| Situation    | Train Error | Test Error | Problem     |
| ------------ | ----------- | ---------- | ----------- |
| Underfitting | High        | High       | Too simple  |
| Overfitting  | Low         | High       | Too complex |
| Good fit     | Low         | Low        | ✅ Ideal    |


![image.png](Images/overfitting_underfitting.jpg)

### How this relates to Linear Regression

Linear Regression often underfits because it can only draw a straight line

So if data is curved: <br>
Code is correct <br>
Model choice is wrong <br>

### How do we FIX these?

Fix underfitting

- Add features

- Use polynomial features

- Use a more powerful model

Fix overfitting

- Simpler model

- More data

- Regularization

## Bias–Variance Tradeoff

### What is Bias?

Bias = error from wrong assumptions

- Model is too simple

- Misses the true pattern

- Strong assumptions about data

📌 Example:

- Using a straight line for curved data

Result:

- ❌ Underfitting

- ❌ High training error

- ❌ High testing error

### What is Variance?

Variance = sensitivity to data

- Model is too complex

- Fits noise instead of pattern

- Changes a lot with small data changes

📌 Example:

- Very wiggly curve passing through every point

Result:

- ❌ Overfitting

- ✅ Low training error

- ❌ High testing error

| Aspect      | High Bias    | High Variance |
| ----------- | ------------ | ------------- |
| Model       | Too simple   | Too complex   |
| Train error | High         | Low           |
| Test error  | High         | High          |
| Problem     | Underfitting | Overfitting   |


### Where Linear Regression fits

- Usually high bias

- Low variance

- Works well only if data is truly linear

So if linear regression performs poorly:

- Code is correct

- Model assumption is wrong

| Problem       | Fix                                      |
| ------------- | ---------------------------------------- |
| High bias     | Add features, polynomial regression      |
| High variance | Simpler model, more data, regularization |


## MSE vs RMSE

### MSE (Mean Squared Error)

- Squares the error

- Punishes big mistakes heavily

- Units are squared (important!)

📌 If price is in lakhs → MSE is in (lakhs)²


### RMSE (Root Mean Squared Error)

RMSE = sqrt(MSE)

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test, y_pred)
```

### Why do we even need RMSE?

🔴 Problem with MSE

- Hard to interpret

- Squared units don’t feel “real”

🟢 RMSE fixes this

- Same unit as target

- Much more intuitive

📌 If:

- RMSE = 2 <br>
👉 “Model is off by ~2 units on average”

| Aspect                | MSE      | RMSE     |
| --------------------- | -------- | -------- |
| Penalizes big errors  | ✅ Strong | ✅ Strong |
| Unit                  | Squared  | Original |
| Interpretability      | ❌ Hard   | ✅ Easy   |
| Used for optimization | ✅ Yes    | ❌ No     |
| Used for reporting    | ❌ Rare   | ✅ Common |


## Polynomial Regression

### The problem Linear Regression can’t solve

Linear Regression can only do this: <br>
y = m x + c

So predictions are always a straight line.

But real data often looks like this 👇

<img src="Images/linear_regression_problem.png" style="background-color: white;" height="300px" width="300px">

A straight line here = underfitting.

### What Polynomial Regression ACTUALLY is

Polynomial Regression is still Linear Regression <br>
We change the FEATURES, not the model

### How do we “make curves” without a new model?

Instead of only x, we add powers of x.

Original feature
x

Polynomial features
x, x², x³

So the model becomes:
y = ax + bx<sup>2</sup> + cx<sup>3</sup> + d

📌 This creates curves, even though the model is still linear in parameters.

### Why is it still called “linear” then?

Because:

- Linear in coefficients (a, b, c)

- NOT linear in x

### Where overfitting enters

Higher degree = more flexibility.

| Degree (power of x)    | Behavior                  |
| --------- | ------------------------- |
| 1         | Straight line (high bias) |
| 2–3       | Usually good              |
| Very high | Overfitting               |


Polynomial regression can overfit VERY easily.
