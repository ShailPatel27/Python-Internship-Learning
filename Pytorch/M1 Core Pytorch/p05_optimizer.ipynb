- model makes a prediction
- loss tells how wrong it is
- gradients tell direction

Now the final question:

> Who actually changes the weights and bias?

Answer: the optimizer.

### What is an optimizer?

An optimizer is just:
- a rule
- that updates weights
- using gradients

Think:

> “Gradient tells where to go, optimizer decides how to move.”

### Very important idea

Imagine weights as knobs

Training loop:
- prediction is bad
- loss is high
- gradient says: “turn knob left”
- optimizer turns it a little
- repeat

After many small turns → good prediction.

### The simplest optimizer: SGD

SGD = Stochastic Gradient Descent

Name sounds scary. Reality is simple.

Rule:
- move weights slightly opposite to gradient

Why opposite?
Because gradient points to more loss.

### Learning rate

Learning rate = step size

- too big → model overshoots, unstable
- too small → model learns very slowly

Think:
- learning rate = how hard you turn the knob

### SGD in PyTorch

`import torch.optim as optim`

`optimizer = optim.SGD(model.parameters(), lr=0.01)`

This means:
- optimize all model weights
- use SGD
- learning rate = 0.01

### The full training step

This is the core loop used everywhere.

 1. forward pass <br>
`y_pred = model(x)`

 2. compute loss <br>
`loss = loss_fn(y_pred, y_true)`

 3. backward pass <br>
`loss.backward()`

 4. update weights <br>
`optimizer.step()`

 5. reset gradients <br>
`optimizer.zero_grad()`


- forward → guess
- loss → how wrong
- backward → how to change
- step → change weights
- zero_grad → clean slate

This is learning.

### Why zero_grad() is mandatory

Gradients accumulate.

If you don’t reset:
- old gradients add up
- updates become wrong

So every step:
- compute
- update
- reset

Always.

### Another optimizer: Adam

Later we'll see:

`optim.Adam(model.parameters(), lr=0.001)`

Difference:
- Adam adapts step size automatically
- works better in most cases

For now:
- SGD = concept clarity
- Adam = practical default

### Long story short

> Optimizer changes weights using gradients to reduce loss.

That’s the entire training story.

- model → makes prediction
- loss → measures error
- gradient → gives direction
- optimizer → updates weights

Repeat this loop → model improves.
