We’ll train a model to learn a very simple rule:

input → output is double the number

So:
- 1 → 2
- 2 → 4
- 3 → 6

The model does not know this rule.
It must discover it.

### Step 1: Create fake data

```python
import torch

#inputs
x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])

#correct outputs
y_true = torch.tensor([[1.0], [4.0], [6.0], [8.0]])
```

### Step 2: Define a tiny model

```python
import torch.nn as nn

class SimpleNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)
        
    def forward(self, x):
        return self.linear(x)
    
model = SimpleNet()
```

Internally:
- one weight
- one bias
- both random

### Step 3: Loss function + optimizer

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

Meaning:
- MSE measures how wrong
- SGD fixes weights
- learning rate = how fast

### Step 4: Training loop

```python
for epoch in range(20):

    # 1. prediction
    y_pred = model(x)

    # 2. MSE
    loss = loss_fn(y_pred, y_true)

    # 3. figures out how y depends on x
    loss.backward()

    # 4. go opposite to gradient
    optimizer.step()

    # 5. reset gradients
    optimizer.zero_grad()

    print(f"Epoch {epoch}: Loss = {loss.item():.4f}")
```

### Step 5: Test the trained model

```python
test_input = torch.tensor([[5.0]])
model(test_input)
```

### What just happened

- model guessed randomly
- loss said “very wrong”
- gradients showed direction
- optimizer adjusted weight & bias
- repeat → better guesses

No intelligence.
No thinking.
Just math + repetition.
