### Integration & Differentiation

Differentiation and Integration are two mathematical operations used to find change in a function or a quantity with respect to another quantity instantaneously and over a period, respectively. <br>
- Differentiation is an instantaneous rate of change, and it breaks down the function for that instant with respect to a particular quantity;
- Whereas Integration is the average rate of change that causes the summation of continuous data of a function over the given period or range. 

Both are the inverse of each other.

### What is Autograd?

Autograd = automatic differentiation

Meaning:
- PyTorch automatically computes gradients
- You don’t manually differentiate formulas

In simple terms:

“PyTorch tracks operations and computes $\frac{d(\text{output})}{d(\text{input})}$ for you.”

#### 1. Gradients (Intuition First)

Gradient = how much output changes when input changes

Example idea:

If
$y = x^2$

Then
$\frac{dy}{dx} = 2x$

(here d is for differentiation)

At $x = 3$:

$\frac{dy}{dx} = 6$

PyTorch does this automatically.

#### 2. requires_grad=True

To tell PyTorch:

“I want to learn from this tensor”

```python
import torch

x = torch.tensor(3.0, requires_grad=True)
y = x**2

display(y)
```

- PyTorch built a computation graph
- No gradient is computed yet

#### 3. Backward Pass (.backward())

```python
y.backward()
```

- walks backwards
- figures out how y depends on x

Now PyTorch computes:

$\frac{dy}{dx}$

Access it using:

```python
x.grad
```

It means: “Show me the change-speed of y with respect to x”

#### 4. Computation Graph

Every operation creates a node.

Example:

```python
x = torch.tensor(2.0, requires_grad=True)

y = x**3
z = y**2
```

Graph conceptually:
- $x \rightarrow y \rightarrow z$
- PyTorch stores how z was created

When you call:

```python
z.backward()
```

```python
x.grad
```

#### 5. Multiple Inputs

```python
x = torch.tensor(2.0, requires_grad=True)
w = torch.tensor(5.0, requires_grad=True)

y = x * w

y.backward()

display(x.grad, w.grad)
```

Math:

$y = xw$

Gradients:
- $\frac{\partial y}{\partial x} = w$

- $\frac{\partial y}{\partial w} = x$

At values:
- $x = 2$
- $w = 3$

Gradients:
- $x.grad = 3$
- $w.grad = 2$

#### 6. Accumulation of Gradients

Gradients add up by default.

```python
x = torch.tensor(3.0, requires_grad=True)

y1 = x ** 2
y1.backward()   # 6

y2 = x ** 3
y2.backward()   # 27

x.grad  # 27 + 6 = 33
```

This gives:
$\frac{d(x^2)}{dx} + \frac{d(x^3)}{dx}$

That’s why we later reset gradients during training.

#### 7. Turning Off Autograd

Sometimes you don’t want gradients (evaluation mode).

- `requires_grad=False` → this tensor will never learn
- `torch.no_grad()` → temporarily stop learning

```python
with torch.no_grad():
    y = x*2
```

Used during:
- validation
- inference
- prediction

#### 8. Key Mental Model

Training = repeat this loop:
- forward pass
- compute loss
- backward pass
- update weights

Autograd handles the hard math part.
