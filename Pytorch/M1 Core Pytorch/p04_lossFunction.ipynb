If a model gives an output…
how does it know whether it’s good or bad?

That’s what loss is.

### What is loss?

Loss = a number that says “how wrong the model is”

- small loss → good prediction
- big loss → bad prediction

#### Example

Suppose:
- correct answer = 10
- model predicted = 7

Question:

How wrong is the model?

One simple way:
$|10 - 7| = 3$

That 3 is the loss.

### Why loss is REQUIRED for learning

A model cannot improve unless:

- it knows it’s wrong
- it knows how wrong
- it can reduce that wrongness

Loss provides direction.

No loss = no learning.

### Loss in neural networks

Most common loss for numbers:

Mean Squared Error (MSE)

Formula (don’t memorize):

$\text{Loss} = (y_{true} - y_{pred})^2$

Why square?
- no negative values
- big mistakes hurt more

### Loss with PyTorch

```python
import torch
import torch.nn as nn

loss_fn = nn.MSELoss()

y_true = torch.tensor([10.0])
y_pred = torch.tensor([7.0])

loss = loss_fn(y_pred, y_true)

loss
```

Because:
$(10 - 7)^2 = 9$

### Where loss fits in training

Training loop logic:

- model predicts
- loss measures error
- gradients tell how to change weights
- weights update
- loss decreases

Loss is the judge.

### How loss + gradients work together

Loss answers:

“How bad is this prediction?”

Gradient answers:

“Which direction should I move to be less bad?”

Together:
- loss = distance
- gradient = direction

This is why differentiation matters.

### Back to weights & bias

Earlier confusion:

“How can just weight and bias learn anything?”

Now answer:

- weights & bias affect prediction
- prediction affects loss
- loss produces gradients
- gradients update weights & bias

Repeat this thousands of times.

That’s learning.

"Training = change weights so loss becomes smaller."
