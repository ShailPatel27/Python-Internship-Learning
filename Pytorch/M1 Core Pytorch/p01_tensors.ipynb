### What is a Tensor?

#### 1. Creating Tensors

- A tensor is the basic data structure in PyTorch.
- AI models only understand tensors, not raw Python objects
- Tensors generalize scalars, vectors, and matrices
<br>
<br>
- 0D tensor → scalar
- 1D tensor → vector
- 2D tensor → matrix
- 3D+ tensor → images, batches, sequences

```python
import torch

a = torch.tensor([1, 2, 3])

b = torch.zeros(3)
c = torch.ones(3)

d = torch.rand(3)

display(a, b, c, d)
```

- torch.tensor() infers datatype automatically
- torch.rand() gives values in range $[0, 1)$

#### 2. Tensor Data Types (dtype)

```python
x = torch.tensor([1, 2, 3])

y = torch.tensor([1, 2, 3], dtype=torch.float32)

display(x.dtype, y.dtype)
```

Common dtypes:
- torch.int64 → integers (default for whole numbers)
- torch.float32 → most used in deep learning

Why floats?
Because gradients are real numbers, not integers.

#### 3. Tensor Shape & Dimensions

```python
t = torch.tensor([
            [1, 2, 3],
            [4, 5, 6]
])

display(t.shape)
```

Meaning:
- 2 rows
- 3 columns

Rank (dimensions):
- 1D → vector
- 2D → matrix
- 3D+ → tensors

#### 4. Reshaping Tensors
(same as numpy)

```python
x = torch.arange(6)
display(x)
```

```python
x_reshaped = x.reshape(2, 3)
display(x_reshaped)
```

- Total number of elements must match
- -1 lets PyTorch infer size automatically

```python
x.reshape(-1, 2)
```

#### 5. Tensor Operations
(same as numpy)

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([10, 20, 30])

display(a+b)
display(a*b)
```

##### Matrix multiplication:

```python
m1 = torch.rand(2, 3)
m2 = torch.rand(3, 2)

display(m1, m2)

torch.matmul(m1, m2)
```

#### 6. Tensor vs NumPy

```python
import numpy as np

np_arr = np.array([1, 2, 3])

torch_arr = torch.from_numpy(np_arr)

display(np_arr)
display(torch_arr)
```

Key points:
- Memory is shared
- Changing one affects the other

```python
np_arr[0] = 99

display(torch_arr)
```
