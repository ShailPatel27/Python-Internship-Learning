### What a neural network really is

A neural network is just a function.

Input → some math → output

Example idea:
- input: number
- output: another number

Inside the network:
- multiply
- add
- apply a simple rule

### The most basic building block — nn.Linear

```python
import torch
import torch.nn as nn
```

```python
layer = nn.Linear(in_features=1, out_features=1)
```

What this means:

- input size = 1
- output size = 1
- internally PyTorch creates:
- a weight
- a bias

Mathematically:

$y = wx + b$

- $x$ → input data (the number you feed into the model)
- $w$ → weight (how important the input is)
- $b$ → bias (shifts the output up or down)

### Passing data through the layer (forward pass)

```python
x = torch.tensor([3.0])
y = layer(x)

print(y)
```

What happened:

- input went in
- PyTorch applied $wx + b$
- output came out

At this stage:
- weights are random
- output is meaningless
- this is normal

### What is nn.Module and why we need it

When networks grow bigger, we need:
- structure
- reusability
- automatic parameter tracking

That’s why we use nn.Module.

### Define your own neural network

```python
class SimpleNet(nn.Module):     #inherit nn.Module
    def __init__(self):
        super().__init__()      #inherit constructor of parent class (nn.Module)
        self.linear = nn.Linear(1, 1)
        
    def forward(self, x):
        return self.linear(x)
```

- SimpleNet → your model
- __init__ → define layers
- self.linear → one linear layer
- forward → how data flows

This is the model.

### Create and use the model

```python
model = SimpleNet()

x = torch.tensor([3.0])
y = model(x)

print(y)
```

- calling model(x) automatically calls forward
- this is a PyTorch feature, not Python

### Where are the weights?

```python
for param in model.parameters():
    print(param)
```

You’ll see:
- weight
- bias

Both have:
- requires_grad=True
- meaning: they can learn

### Add an activation function (optional but important)

Without activation:
- model is just linear

Add ReLU (Rectified Linear Unit):

```python
class SimpleNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.linear(x)
        y = self.relu(x)
        
        return x, y
    
model = SimpleNet()
x = torch.tensor([3.0])
y = model(x)
print(y)
```

Now:
- negative values → 0
- model becomes non-linear

#### Key mental model

- nn.Linear → does math
- forward() → defines flow
- model parameters → learnable numbers
- autograd → updates them
