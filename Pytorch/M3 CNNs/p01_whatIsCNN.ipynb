### What is a CNN

A Convolutional Neural Network (CNN) is a type of deep learning algorithm designed to analyze, classify, and recognize visual data like images and videos

### What problem CNNs solve

Tabular data:
- each column is independent

Images:
- pixels are spatially related
- nearby pixels matter together

Fully-connected layers (nn.Linear) ignore spatial structure. <br>
CNNs exploit spatial structure.

### One image = just a tensor

An image in PyTorch is:

`[channels, height, width]`


Example:
- grayscale → `[1, 28, 28]`
- RGB → `[3, 224, 224]`

A batch of images:

`[batch_size, channels, height, width]`


This is the only shape change you need to care about.

### What changes in the model

##### Before (tabular)
`nn.Linear(in_features, out_features)`

##### Now (images)
`nn.Conv2d(in_channels, out_channels, kernel_size)`

That’s it.

Everything else stays the same.

### What Conv2d does

Instead of:
- looking at all inputs at once

It:
- looks at small patches
- slides across the image
- learns patterns like:
- edges
- corners
- textures

You don’t code these patterns - the model learns them.

### Mental mapping

| Tabular DL          | CNN                |
| ------------------- | ------------------ |
| Feature             | Pixel              |
| `nn.Linear`         | `nn.Conv2d`        |
| Feature interaction | Spatial pattern    |
| Same training loop  | Same training loop |
