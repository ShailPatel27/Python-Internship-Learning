### MNIST

MNIST Stands for Modified National Institute of Standards and Technology.

It is a dataset comprising 70,000 grayscale, 28 X 28 pixel images of handwritten digits (0-9).

It consists of 60,000 training and 10,000 testing images

<img src="images/MNIST_dataset_example.png">

### Why MNIST

We’ll use MNIST because:
- tiny
- fast
- classic
- perfect for first CNNs

### Step 1. Load the dataset

```python
import torch
from torchvision import datasets, transforms
```

```python
transform = transforms.ToTensor()
```

```python
train_data = datasets.MNIST(
    root="data",
    train=True,
    download=True,
    transform=transform
)

test_data = datasets.MNIST(
    root="data",
    train=False,
    download=True,
    transform=transform
)
```

What this does:
- downloads images
- converts them to tensors
- normalizes pixel values to [0, 1]

### Step 2. Wrap with DataLoader

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64)
```

### Step 3. Inspect one batch

```python
images, labels = next(iter(train_loader))
images.shape, labels.shape
```

Interpretation:
- 64: images per batch
- 1: channel - grayscale     (channel=3 is RGB)
- 28×28: pixels (size of each image)

### Step 4. Visualize images

```python
import matplotlib.pyplot as plt

plt.imshow(images[0].squeeze(), cmap="gray")
plt.title(f"Label: {labels[0].item()}")
plt.axis("off")
plt.show()
```
