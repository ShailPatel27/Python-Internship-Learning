### The uncomfortable truth about real-world CNNs

- almost nobody trains CNNs from scratch
- even large companies reuse pretrained models
- training from scratch is mostly for research or learning

Why
- training deep CNNs is expensive
- requires massive datasets
- requires weeks of compute

So the industry solution is transfer learning.

### What transfer learning means

Transfer learning means

- take a CNN trained on a huge dataset
- reuse its learned features
- adapt it to your smaller task

Think of it like this

- early layers learn edges and textures
- middle layers learn shapes and parts
- later layers learn task-specific concepts

We reuse the first two.
We replace the last part.

### Why this works

CNNs trained on large datasets (like ImageNet) learn

- edge detectors
- color gradients
- texture patterns
- shape primitives

These are universal visual features.

So even if your task is
- cats vs dogs
- cars vs trucks
- medical images

Those early features still apply.

### What ImageNet is

ImageNet is

- 14+ million images
- 1000 object classes
- massive visual diversity

CNNs trained on ImageNet have seen

- animals
- vehicles
- textures
- backgrounds
- lighting variations

This is far beyond CIFAR-10.

### What a pretrained CNN looks like conceptually

A pretrained CNN is not one thing — it’s layers stacked like this:

<pre>
[Edge detectors]
      ↓
[Texture detectors]
      ↓
[Shape detectors]
      ↓
[Object-level concepts]
      ↓
[Classifier]
</pre>


Transfer learning keeps
- everything above the classifier

And replaces
- the classifier

### Common pretrained CNNs you’ll hear about

Names you will encounter

- ResNet
- VGG
- MobileNet
- EfficientNet

They differ in
- depth
- speed
- memory usage

But they all follow the same idea:

> learn general vision features once, reuse everywhere.
