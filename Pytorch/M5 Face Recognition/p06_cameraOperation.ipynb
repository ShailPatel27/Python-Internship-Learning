### Important Concepts

#### 1. OpenCV Frame Format

OpenCV gives:

`BGR image (numpy array)`

But MTCNN expects:

`PIL image (RGB)`

So we must convert.

#### 2. Processing Speed

We don’t want to:

- Recompute database every frame

- Reinitialize model every frame

Database and models must be loaded once.

#### 3. Continuous Loop

We use:

`while True:`

And break when key pressed.

### Webcam Capture

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    cv2.imshow("Webcam", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

#### What is `VideoCapture`?

It is an OpenCV class that:

> Connects to a video source (camera or file)

Values:

- 0 → default webcam

- 1 → second camera (if exists)

- "video.mp4" → video file

So:

`cap = cv2.VideoCapture(0)`

means:

“Open my default webcam”

#### `ret, frame = cap.read()`

`cap.read()` returns TWO things:

$1.$ `ret`(boolean)

- True → frame captured successfully

- False → camera failed / ended

$2.$ `frame`

- A numpy array

- Shape: (height, width, 3)

- Format: BGR (not RGB)

Example shape:

`(480, 640, 3)`

That means:

- 480 pixels tall

- 640 pixels wide

- 3 color channels

#### `cv2.imshow("Webcam", frame)`

This displays an image in a window.

Arguments:

`cv2.imshow(window_name, image)`

- "Webcam" → title of the window

- frame → the image to show

#### `cv2.waitKey(1)`

This waits for a keyboard input.

Important:

`cv2.waitKey(1)`

Means:

> “Wait 1 millisecond for key press”

If we wrote:

`cv2.waitKey(0)`

It would freeze until a key is pressed.

`if cv2.waitKey(1) & 0xFF == ord('q'):`

- waitKey(1) → returns integer key code

- & 0xFF → ensures correct format across systems

- ord('q') → ASCII value of letter q

So this means:

If user presses q → exit loop

#### Clean Exit

`cap.release()`

- Releases camera resource.

`cv2.destroyAllWindows()`

- Closes all OpenCV windows.

If you don’t do this, camera may stay locked.
