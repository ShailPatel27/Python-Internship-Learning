```python
import urllib.request
import zipfile
import os

os.makedirs("models", exist_ok=True)

print("Downloading...")
urllib.request.urlretrieve(
    "https://zenodo.org/records/14257640/files/gaze-estimation-main.zip?download=1",
    "models/gaze-estimation-main.zip"
)

print("Extracting...")
with zipfile.ZipFile("models/gaze-estimation-main.zip", "r") as z:
    z.extractall("models/")

print("Done")
```

```python
from uniface import MobileGaze
import cv2

gaze_model = MobileGaze()
frame  = cv2.imread("../../M5 Face Recognition/faces/single/shail/shail1.jpg")
result = gaze_model.estimate(frame)
print(dir(result))
print(result)
```

```python
from uniface import MobileGaze
import cv2

gaze_model = MobileGaze()
cap = cv2.VideoCapture(0)

ret, frame = cap.read()
if ret:
    result = gaze_model.estimate(frame)
    print(dir(result))
    print(result)

cap.release()
```
