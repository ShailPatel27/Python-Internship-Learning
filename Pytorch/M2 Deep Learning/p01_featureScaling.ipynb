### What is feature scaling?

Feature scaling means:
- putting all input features on a similar numeric range

Example (real problem):

- feature A (income) → 10,000 – 1,000,000
- feature B (rooms) → 1 – 10

Neural network sees:
- income as far more important
- even if it’s not

That breaks learning.

### Why this matters specifically for neural networks

Neural networks learn by:
- multiplying inputs by weights
- computing gradients

Large input values:
- create huge gradients
- cause unstable updates

Small input values:
- create tiny gradients
- slow learning

So learning becomes:
- noisy
- slow
- or completely broken

Then:
- first feature dominates
- gradients explode
- optimizer struggles

### The solution: Scaling

We transform inputs so their values are close to each other

Now:
- gradients behave
- loss decreases smoothly
- learning stabilizes

### Scaling methods

#### 1. Standardization (most common for NN)

Formula:

$\frac{x - \mu}{\sigma}$

Meaning:
- center data around 0
- spread evenly

Result:
- mean ≈ 0
- std ≈ 1

This is default for tabular neural networks.

#### 2. Min-Max scaling

Formula:

$\frac{x - x_{min}}{x_{max} - x_{min}}$

Result:
- values between 0 and 1

Used when:
- inputs have strict bounds

### Pipeline:

- raw data
- scale features
- convert to tensors
- Dataset
- DataLoader
- model

### Example

```python
from sklearn.datasets import fetch_california_housing
X = fetch_california_housing().data
print(X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Now:
- X_scaled is safe for neural networks

Then:

```python
import torch
X_scaled = torch.tensor(X_scaled, dtype=torch.float32)
print("X", X)
print("X_scaled", X_scaled)
```
