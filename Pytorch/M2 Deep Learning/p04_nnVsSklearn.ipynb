Neural Network vs Scikit-Learn (Same Data, Same Task)

### Step 1 – Use a real, familiar dataset

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = fetch_california_housing()

X, y = data.data, data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Step 2 – Scikit-Learn baseline

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_mse = mean_squared_error(y_test, lr_pred)

print("Linear Regression MSE:", lr_mse)
```

### Step 3 – Same task with PyTorch

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)

X_test_t = torch.tensor(X_test, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
```

```python
train_ds = TensorDataset(X_train_t, y_train_t)
test_ds = TensorDataset(X_test_t, y_test_t)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
```

### Step 4 – Neural network (small & fair)

```python
model = nn.Sequential(
    nn.Linear(X.shape[1], 64),
    nn.ReLU(),
    nn.Linear(64, 1)
)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

### Step 5 – Train NN

```python
for epoch in range(10):
    for xb, yb in train_loader:
        pred = model(xb)
        loss = loss_fn(pred, yb)
        
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
print(loss)
```

You will see ONE of these:

Case 1 NN ≈ Linear Regression

✔ NN not needed <br>
✔ Scikit-Learn wins (simpler, faster) <br>

Case 2 NN slightly better

✔ NN captures non-linearity <br>
✔ Worth using NN <br>

Case 3 NN worse

❌ Overkill <br>
❌ Needs tuning <br>
❌ Stick to classical ML <br>
