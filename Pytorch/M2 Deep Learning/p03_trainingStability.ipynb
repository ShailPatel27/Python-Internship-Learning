This module answers:

“Why did my loss suddenly become nan?”
“Why was it working yesterday and broken today?”

### The 3 knobs that control stability

Almost all training problems come from these:

- learning rate
- batch size
- initialization

#### 1. Learning rate

Learning rate = how big a step the optimizer takes

Think:
- too big → jump past the solution
- too small → crawl forever

High Learning Rate:

- loss jumps up and down
- loss becomes inf or nan
- training feels chaotic
- optimizer overshoots the minimum


Low Learning Rate:

- loss decreases extremely slowly
- looks “stuck”
- training technically works, but useless

##### Safe defaults (tabular data)

- Adam → lr = 0.001
- SGD → lr = 0.01 or 0.001

#### 2. Batch size (stability vs noise)

Batch size controls:
- how noisy gradients are

Small batch size

- noisy loss curve
- more randomness
- can generalize better

Large batch size

- smooth loss curve
- more stable updates
- needs more memory

#### 3. Why loss sometimes becomes nan

Almost always due to:
- unscaled features
- learning rate too high
- exploding gradients

We already fixed:
- scaling

So next fix is:
- reduce learning rate

### Debug checklist

If training breaks:

- check feature scaling
- lower learning rate
- print loss every epoch
- verify shapes
- ensure .zero_grad() is called

90% of issues die here.

### Gradient clipping

`torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)`

This:
- prevents extreme updates
- used in RNNs & unstable models

Not needed daily - just know it exists.
