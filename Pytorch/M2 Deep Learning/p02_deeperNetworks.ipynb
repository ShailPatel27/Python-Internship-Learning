### What does “deeper” mean?

Depth = number of layers, not neurons.

- shallow network → 1 hidden layer
- deep network → multiple hidden layers

Example:

Input → Linear → Output        (shallow) <br>
Input → Linear → ReLU → Linear → ReLU → Output   (deep)


Same training loop. Same loss. Same optimizer.

Only more layers.

### Why more layers help

Each layer learns a transformation.

Think like this:
- layer 1 → simple patterns
- layer 2 → combinations of patterns
- layer 3 → higher-level structure

Instead of learning:

> “price directly from raw features”

The model learns:

> features → representations → decision


Depth ≠ intelligence <br>
Depth = capacity

More layers mean:
- model can represent more complex relationships
- but also easier to mess up

### Example: shallow vs deep (same task)

#### Shallow model:

```python
import torch.nn as nn

nn.Linear(8, 1)
```

#### Deep model:

```python
nn.Sequential(
    nn.Linear(8, 64),
    nn.ReLU(),
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 1)
)
```

What changed:
- nothing in training code
- only the model definition

### When deeper networks HELP

- data has complex relationships
- interactions between features matter
- linear models underperform

Examples:
- housing prices
- user behavior
- financial risk

### When deeper networks HURT

- very small datasets
- simple linear relationships
- no regularization

Symptoms:
- training loss ↓
- validation loss ↑

That’s overfitting.
