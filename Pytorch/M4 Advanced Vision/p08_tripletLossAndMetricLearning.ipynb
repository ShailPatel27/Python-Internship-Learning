### What we are trying to learn

We want a function:

$f(x) = v$


Where:

- $v \in \mathbb{R}^d$
- similar images → close vectors
- different images → far vectors

Not classification.

We are shaping geometry.

### The Triplet Structure

Each training sample consists of:

- Anchor (a)
- Positive (p) → same identity as anchor
- Negative (n) → different identity

So instead of one image per sample,
we now feed three.

### What the loss enforces

We want:

$d(a, p) < d(a, n)$


More precisely:

$d(a, p) + m < d(a, n)$


Where:

- $d$ = distance
- $m$ = margin

If this condition is violated, we apply penalty.

### Triplet Loss formula

> $\mathcal{L} = \max(0, d(a,p) - d(a,n) + m)$
<br>

If:

> $d(a,n)$ is much larger


Loss becomes 0.

If:

> $d(a,p)$ too close to $d(a,n)$


Loss is positive.

### Why margin matters

Margin forces separation.

Without margin:

- Model might collapse embeddings.

With margin:

- We require minimum gap.

Example:

> $m = 0.5$


means negative must be at least 0.5 farther than positive.

### Which distance do we use?

Common:

- Euclidean distance
- Cosine distance

Euclidean:

$d(x,y) = \|x - y\|_2$


Often embeddings are normalized:

$v = \frac{v}{\|v\|}$


Then Euclidean and cosine behave similarly.

### What embedding training does geometrically

Over time:

- same identities cluster
- different identities separate
- embedding space becomes structured

This is called metric learning.

We are learning a space where distance = semantic similarity.

### Why this is better than classification for faces

Classification:

- image → class ID


Embedding:

- image → vector → compare to database


Advantages:

- new identities without retraining
- scalable
- flexible
- open-set recognition

That’s why FaceNet, ArcFace, etc. use metric learning.

### What changes in architecture?

Almost nothing.

Instead of:

- Linear → softmax


We do:

- Linear → embedding vector (e.g., 128-d)


No softmax layer.

Just embeddings.
