## Failure Modes & Bias in Vision Models

Goal
- understand why high accuracy ≠ robustness
- understand shortcut learning
- understand texture bias
- understand distribution shift

This is where real-world ML becomes serious.

### Accuracy is not understanding

You can get:

> 99% test accuracy


And still fail in the real world.

Why?

Because models optimize:

> $\min \mathcal{L}_{train}$


They do not optimize understanding.

They optimize shortcuts.

### Shortcut Learning

Models exploit easiest predictive signals.

Example:

If in training:

- Wolves appear in snow
- Dogs appear on grass

Model may learn:

“Snow → wolf”

Instead of learning animal structure.

This is shortcut learning.

### Texture Bias vs Shape Bias

Research showed:

CNNs rely heavily on texture, not shape.

If you take:

- Elephant shape
- Dog texture

CNN may classify it as dog.

Humans rely more on shape.

CNNs rely more on local patterns.

Why?

Because convolution emphasizes local correlations.

### Distribution Shift

Training distribution:

> $p_{train}(x)$


Real-world distribution:

> $p_{real}(x)$


If:

> $p_{train}(x) \neq p_{real}(x)$


Performance collapses.

This is called domain shift.

Examples:

- Different lighting
- Different camera
- Different background
- Different resolution

Models are fragile outside training distribution.

### Overconfidence

Softmax probabilities are not calibrated.

Model can output:

`99.9% confidence`


On complete nonsense.

Because softmax enforces relative confidence.

Not epistemic uncertainty

### Adversarial Vulnerability

Small perturbation:

$x' = x + \epsilon$


Where:

$\|\epsilon\|$ small


Can cause misclassification.

Why?

Because high-dimensional linear behavior dominates locally.

Deep networks are locally almost linear.

### Why this matters for face recognition

In face systems:

Failures include:

- lighting changes
- occlusion
- pose variations
- demographic bias
- spoof attacks

High benchmark accuracy does not mean safe deployment.

### Big Realization

Deep models:

- are pattern optimizers
- not reasoning engines
- not causal models

They detect statistical correlations.

If those correlations change, performance drops.

### What separates strong engineers from beginners

Beginners:

“Model got 95%. Done.”

Advanced practitioners ask:

- What shortcuts did it learn?
- What distribution assumptions exist?
- What happens under domain shift?
- Is it calibrated?
- Is it biased?

That’s model maturity.
