## Vision Transformers — how attention replaces convolution

Goal
- understand how ViT processes images
- see how patches replace kernels
- understand why attention is global
- compare CNN vs ViT inductive bias

### How CNN processes images

CNN:

- sliding local kernels
- local receptive field
- translation equivariant
- hierarchical feature building

Receptive field grows gradually:

Receptive field = 1 + 2N


CNN has strong inductive bias:
“Local structure matters.”

### How Vision Transformer processes images

ViT does something radical.

Instead of sliding kernels:

- Split image into patches
- Flatten patches
- Treat patches as tokens

Example:

224×224 image <br>
Patch size = 16×16

Number of patches:

> $(224 / 16)^2 = 14^2 = 196$


So image becomes:

> 196 tokens


Just like words in NLP.

### Patch embedding

Each patch:

> 16×16×3


Flatten → vector.

Then projected to embedding dimension:

> $\mathbb{R}^{768}$


Now image is a sequence of vectors.

### Self-Attention mechanism

Attention computes:

$Attention(Q,K,V) = softmax\left(\frac{QK^T}{\sqrt{d}}\right)V$


Where:

- Q = queries
- K = keys
- V = values

Each patch can attend to every other patch.

This means:

Global receptive field immediately.

No gradual growth.

### Key difference from CNN

|CNN|ViT|
|-|-|
|Local → global gradually | Global from start |
|Built-in locality|No locality bias|
|Efficient|Needs more data|

CNN assumes: <br>
 “Nearby pixels are related.”

ViT learns that from data.

### Why ViTs need more data

Because they lack inductive bias.

CNN already assumes:
- translation equivariance
- locality

ViT must learn those patterns.

So small datasets → CNN better
Large datasets → ViT often better

### Why ViT scales well

Attention complexity:

$O(N^2)$


Where N = number of patches.

For moderate patch size, manageable.

ViTs scale well in very large models.

### When CNN still wins

- Small datasets
- Low compute
- Real-time systems
- Edge devices

CNNs are still dominant in many production systems.

### Modern hybrid approach

Many systems combine:

- CNN stem
- Transformer body

Or use Conv-like attention.

Architecture is converging.
