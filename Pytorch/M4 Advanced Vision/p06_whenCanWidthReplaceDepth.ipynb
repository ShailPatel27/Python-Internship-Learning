### The Universal Approximation Theorem

Mathematically:

A single hidden-layer neural network with enough width can approximate any continuous function.

Formally:

$f(x) \approx W_2 \sigma(W_1 x)$


With enough neurons in that hidden layer,
it can approximate arbitrary mappings.

So in theory:

Depth is not required.

### Then why do we use deep networks?

Because theory ≠ efficiency.

A shallow network might require:

	millions or billions of neurons


to represent what a deep network can do compactly.

Depth allows compositional reuse:

$f(x) = f_3(f_2(f_1(x)))$


Each stage builds on previous structure.

### Concrete intuition

Suppose you want to detect:

A square inside a circle inside a scene.

Shallow wide network:

- must directly learn mapping from pixels → full pattern.

Deep network:

- layer 1 → detect edges

- layer 2 → detect shapes

- layer 3 → detect square

- layer 4 → detect circle

- layer 5 → detect composition

Depth mirrors structure of reality.

### Why width sometimes helps

Very wide networks:

- increase representational capacity

- reduce gradient noise

- sometimes generalize better

Example:

- Wide ResNets outperform very deep thin ResNets.

So width can compensate somewhat.

### Why width cannot fully replace depth

Depth increases:-

Receptive field growth:

`Receptive field = 1 + 2N`


Width does not change receptive field.

If a neuron cannot see entire object, <br>
no amount of width helps.

Width = more parallel filters <br>
Depth = larger spatial reasoning

Different roles.

### Computational difference

Wide shallow network:

- heavy matrix multiplications

- huge parameter count

Deep narrow network:

- fewer parameters

- more sequential operations

Modern models balance both.

### Practical conclusion

Width helps when:

- network is under-parameterized

- gradients are noisy

- training is unstable

Depth helps when:

- task requires hierarchical structure

- objects are large

- abstraction is required

Best models use both.
