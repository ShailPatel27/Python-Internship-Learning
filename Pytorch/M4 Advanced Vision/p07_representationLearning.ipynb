### 1. What classification actually does

In normal classification, your model outputs:

$z = W h + b$


Where:

- $h$ = final feature vector
- $W$ = classifier weights
- $z$ = logits

Then softmax:

$\hat{y}_i = \frac{e^{z_i}}{\sum_j e^{z_j}}$


This forces the model to:

- separate fixed classes
- learn decision boundaries
- assign probability to known labels

Problem:
- it only works for predefined classes

### 2. The limitation of classification

Suppose you train a model on:

- 10 people

Now a new person appears.

Softmax cannot handle:

- unseen identities
- open-set recognition
- similarity search

It only answers:

“Which of my known classes is this?”

Not:

“Is this the same person?”

### 3. What representation learning does instead

Instead of predicting labels,
we learn a function:

$f(x) = v$


Where:

- $v \in \mathbb{R}^d$
- $v$ is an embedding vector

Now classification is replaced by:

Distance in embedding space.

### 4. What is an embedding?

An embedding is:

A learned vector representation where:

- similar inputs → close vectors
- different inputs → far vectors

Similarity is often measured by:

Cosine similarity:

$\text{cos}(v_1, v_2) = \frac{v_1 \cdot v_2}{\|v_1\|\|v_2\|}$


Or Euclidean distance:

$\|v_1 - v_2\|_2$

### 5. Why this is powerful

With embeddings:

- You can compare any two images
- You don’t need fixed classes
- You can add new identities without retraining

This is how:

- Face recognition works
- Image retrieval works
- Recommendation systems work

### 6. Conceptual Difference

Classification model learns:

`image → class ID`


Embedding model learns:

`image → position in feature space`


One is discrete. <br>
One is geometric.

### 7. What changes in training?

Instead of minimizing cross entropy:

$\mathcal{L}_{CE}$


We use metric losses like:

Contrastive loss:

$\mathcal{L} = y \|v_1 - v_2\|^2 + (1-y) \max(0, m - \|v_1 - v_2\|)^2$


Triplet loss:

$\mathcal{L} = \max(0, d(a,p) - d(a,n) + m)$


Where:

- $a$ = anchor
- $p$ = positive (same identity)
- $n$ = negative (different identity)
- $m$ = margin

This pushes:

- same identities together
- different identities apart

### 8. Why this connects to face recognition

Face recognition is NOT classification.

It is:

`image → embedding → compare distances`


If distance < threshold → same person.

That’s it.

No softmax required.
