### Why plain deep networks struggle

Consider a deep network:

$y = f_n(f_{n-1}(...f_1(x)))$


- During backpropagation, gradients multiply:

$\frac{\partial y}{\partial x}
= \prod_{i=1}^{n} \frac{\partial f_i}{\partial x}$


- This is the key.

If each derivative is slightly < 1:

Example:

$0.9^{50} \approx 0.005$


- Gradient becomes tiny.

- If slightly > 1:

$1.1^{50} \approx 117$


- Gradient explodes.

- So deep networks become unstable.

This is not about capacity. <br>
It is about multiplication of small numbers.

### What residual connections change

Residual block:

$y = x + g(x)$


Now compute gradient:

$\frac{\partial y}{\partial x}
= 1 + \frac{\partial g(x)}{\partial x}$


Notice the difference:

Plain network → product
Residual network → sum with 1

That constant 1 guarantees:

- gradient can’t vanish completely
- signal always has a direct path
- optimization becomes stable

This is the entire breakthrough.

### Why this enables very deep networks

Without residuals:
- 20–30 layers often unstable

With residuals:
- 50, 100, 150 layers train fine

Because gradient now flows through:

identity path → unblocked


Deep learning became practical at scale because of this.

### Why Transformers copied this idea

Look at a Transformer block:

`x = x + Attention(x)` 	#Computes a transformation of x<br>
`x = x + MLP(x)`


That’s residual learning.

Transformers would collapse without skip connections.

So residual connections are not “a CNN trick”.

They are an optimization principle.

### Where is this implemented

When you see modern architectures:

- ResNet

- Vision Transformer
- GPT
- Diffusion models

They all use:

$y = x + g(x)$
