## ArcFace & Angular Margin Loss

Goal
- understand why triplet loss is inefficient
- understand angular margin
- see why ArcFace scales better
- understand why modern face systems use it

### The problem with triplet loss

Triplet loss:

$\mathcal{L} = \max(0, d(a,p) - d(a,n) + m)$


Problems:

- requires careful triplet mining
- slow convergence
- unstable early training
- difficult with millions of identities

Triplet loss is geometrically correct but operationally messy.

### A smarter idea

Instead of:

> “Push embeddings apart using triplets”

ArcFace says:

> “Use classification, but modify geometry.”

### Key observation

In classification, logits are:

$z_i = W_i^T h + b_i$


If we normalize:

$\|W_i\| = 1$ <br>
$\|h\| = 1$


Then:

$z_i = \cos(\theta_i)$


Where:

- $\theta_i$ = angle between embedding and class weight

Now classification becomes angular separation.

### What ArcFace changes

Instead of normal softmax:

$\cos(\theta_y)$


ArcFace modifies the target logit:

$\cos(\theta_y + m)$


Where:

- $m$ = angular margin

Final loss:

$\mathcal{L} = -\log \frac{e^{s \cos(\theta_y + m)}}{e^{s \cos(\theta_y + m)} + \sum_{j \neq y} e^{s \cos(\theta_j)}}$


Where:

- $s$ = scaling factor

### What this does geometrically

Instead of just separating classes,

It enforces:

- minimum angular gap
- tighter intra-class clusters
- larger inter-class separation

In angular space.

No triplets needed.

### Why angular space is better

Euclidean distance depends on magnitude.

Angular distance depends only on direction.

If embeddings are normalized:

$\|h\| = 1$


Then all embeddings lie on a hypersphere.

This stabilizes geometry.

### Why ArcFace scales better

Triplet loss:

- needs sampling strategy
- O(batch²) comparisons

ArcFace:

- standard classification
- no mining
- stable gradients
- easy scaling to millions of identities

That’s why almost all production face systems use angular margin losses.

### Big conceptual shift

Triplet learning:

`optimize distances directly`


ArcFace:

`optimize classification with geometric constraint`


It merges both worlds.
