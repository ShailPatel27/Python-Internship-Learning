## Width vs Depth — Why They Are Not Equivalent

<img src="images/wide_vs_deep_nn.png">

### What does “depth” mean?

Depth is just the number of weight layers

eg:
- Conv2d
- Linear

nn.ReLU() doesnt count as a layer since it doesn't hold any weight

Depth = number of sequential transformations.

Mathematically:

$y = f_n(f_{n-1}(...f_1(x)))$


Each layer transforms the representation further.

Depth increases:

- receptive field
- abstraction level
- compositional reasoning

### What does “width” mean?

Width is just the number of neurons in a layer

Width = number of channels (feature maps).

Example:

`nn.Conv2d(32, 128, 3)`


Here width = 128 channels.

Wider means:

- more parallel filters
- more feature diversity at same level
- larger representational capacity

### Key conceptual difference

Depth adds hierarchy. <br>
Width adds variety at same level.

Depth:

`edges → textures → shapes → objects`

Width:

`many types of edges simultaneously`

These are not interchangeable.

### Why depth changes function class

A shallow but very wide network:

$y = f(x)$


can approximate complex functions,
but requires huge parameter count.

Deep network:

$y = f_3(f_2(f_1(x)))$


can represent compositional structure efficiently.

Depth enables:

- reuse of features

- hierarchical composition

- exponential growth in representational power

### Concrete intuition

Imagine recognizing a face.

Shallow wide network:

- must detect full face pattern directly.

Deep network:

- detect edges

- combine into eyes

- combine into face

Compositional reasoning.

Depth mirrors real-world structure.

### Computational trade-off

Increasing width:

- increases parameters linearly

- increases memory usage

- increases compute per layer

Increasing depth:

- increases sequential operations

- increases latency

- improves abstraction

Modern models balance both.
