Goal
- understand why “just stacking layers” fails
- understand why deeper ≠ better by default
- build intuition for residual connections (before code)

### The naive assumption (and why it’s wrong)

Naive idea
- “If 5 layers work, 50 layers should work better”

Reality
- deeper CNNs often train worse
- loss stops improving
- accuracy stagnates or drops

This happens even when overfitting is NOT the issue.

So something deeper is broken.

### The real problem is NOT just vanishing gradients

Two real problems exist:

- gradients get weaker
- optimization itself becomes harder

The second one is more important.

### Why deeper networks are harder to optimize

Let’s reason simply.

Each layer tries to learn a function:

$y = f(x)$

Meaning
- input $x$
- layer applies transformation $f$
- output $y$

In deep networks, layers stack:

$y = f_3\big(f_2(f_1(x))\big)$

Meaning
- first layer computes $f_1(x)$
- second layer computes $f_2(\cdot)$
- third layer computes $f_3(\cdot)$
- output is nested composition

This nesting is what makes optimization hard.

This forces the network to learn:
- complex transformations
- even when identity mapping would be enough

Learning “do nothing” is surprisingly hard.

### The identity mapping insight

Sometimes, the best thing a layer can do is:

$f(x) = x$

Meaning
- layer does nothing
- information passes through unchanged

Surprisingly hard for deep networks to learn.

But standard CNN layers:
- are not biased toward identity
- must learn identity from scratch
- often fail to do so

So deeper layers actively make things worse.

### Residual learning: the core idea

Instead of forcing layers to learn:

$f(x)$

We ask them to learn:

$y = x + g(x)$

Where
- $x$ is the input
- $g(x)$ is a small correction

This is much easier to learn.

If the best thing is “do nothing”:
- just learn $g(x)=0$

This single change fixes deep optimization.

### What a residual connection really does

A residual block computes:

$y=x+g(x)$

Key consequences:

- identity path always exists
- gradients flow directly backward
- deeper networks cannot perform worse than shallower ones (in theory)

This is not magic — it’s optimization engineering.

### Why this changed everything

Before residuals:
- 20–30 layers was hard
- 50 layers often failed

After residuals:
- 50 layers worked
- 100+ layers worked
- training became stable

This is why ResNet won ImageNet and changed vision forever.

### Why this matters beyond ResNet

Residual ideas appear in:

- DenseNet
- Transformers (skip connections everywhere)
- Vision Transformers
- Diffusion models

If you understand residuals,<br>
modern deep learning architectures stop feeling random.
