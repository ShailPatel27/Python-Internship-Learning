## Receptive Field: What a CNN Actually Sees

Goal
- understand what “receptive field” means
- see how depth changes spatial awareness
- connect this to why deeper models matter

### What is a receptive field?

The receptive field of a neuron is:

> The region of the input image that affects that neuron’s output.

If a neuron in layer 5 activates, it only depends on a certain patch of the original image.

### Start simple: one convolution

Suppose:

`nn.Conv2d(..., kernel_size=3, padding=1)`


Kernel size = 3.

That means each output pixel depends on:

$3 \times 3$


area of the input.

So receptive field = 3.

### Stack two 3×3 convolutions

Now consider:

`Conv(3×3)` <br>
`Conv(3×3)`


First layer sees:

$3 \times 3$


Second layer sees:

$3 \times 3$ of the previous layer


But each of those already depended on 3×3.

So total receptive field becomes:

$5 \times 5$


General rule (stride=1):

Each new 3×3 layer adds 2 to receptive field.

### After N layers

If kernel size = 3, stride = 1:

`Receptive field = 1 + 2N`


Example:

- 1 layer → 3

- 2 layers → 5

- 5 layers → 11

- 50 layers → 101

So deep networks see much larger context.

### Why this matters

Small receptive field:
- sees edges
- local patterns

Large receptive field:
- sees object parts
- global structure

Without enough depth:

- network cannot “see” entire object
- struggles with large shapes

This is why shallow CNNs fail on complex scenes.

### Pooling changes this faster

If you add:

`nn.MaxPool2d(2)`

Stride=2 halves spatial size.

Now each neuron covers a larger region of the original input.

Pooling increases receptive field faster than plain conv stacking.

### Important subtlety

Theoretical receptive field ≠ effective receptive field.

Even if math says:

$101 \times 101$


In practice, center pixels influence more than corners.

But depth still increases global awareness.

### Why deeper models matter now

If image size is 224×224:

- 5-layer CNN → sees small patches

- 50-layer CNN → sees almost entire image

That allows reasoning about:

- object configuration
- relationships
- large shapes

Depth = spatial reasoning capacity.
