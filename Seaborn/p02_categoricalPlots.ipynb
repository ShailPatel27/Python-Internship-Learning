```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

### countplot — COUNTING CATEGORIES

Histogram but with a different display format

```python
sns.countplot(data=df, x="day")
plt.show()
```

```python
sns.countplot(data=df, x="day", hue="sex")
plt.show()
```

### barplot — AGGREGATED NUMBERS

The bar height is the average of data <br>

The black line is a confidence interval (CI), also called an error bar. <br>
It shows how uncertain the average (mean) value is. <br>

Short black line
- Mean is reliable (less variation)

Long black line

- Mean is less reliable (more variation / fewer samples)

```python
sns.barplot(data=df, x="day", y="total_bill")
plt.show()
```

#### Estimator

estimator is a statistical function used to calculate a single summary value (like the mean, median, or sum) for all the data points within a specific category or bin.

Values for estimator:
- mean (default)
- median
- sum
- len
- std
- min
- max

```python
sns.barplot(data=df, x="day", y="total_bill", estimator="sum")
```

#### Remove confidence interval

By default, Seaborn shows a confidence interval (thin black line).

```python
sns.barplot(data=df, x="day", y="total_bill", errorbar=None)
# sns.barplot(data=df, x="day", y="total_bill", ci=None)  #Depreciated

plt.show()
```

### boxplot — DISTRIBUTION + OUTLIERS

Shows:

- Median
- Spread
- Outliers

```python
sns.boxplot(data=df, x="day", y="total_bill")

# Saturday has higher spread
# Some bills are unusually high (outliers)
```

```python
sns.boxplot(data=df, x="day", y="total_bill", hue="sex")
```

### violinplot – Boxplot + Density

- Shows shape of data distribution.
- It is the same as a boxplot and even has one.
- Along with the box, data density is also shown.
- In the given plot, more data is gathered at the center where the median is.
- The data above and below the box plot shows outliers

```python
sns.violinplot(data=df, x="day", y="total_bill")
```

```python
# violin plot with no outliers

data = {
    "group": ["A"] * 20 + ["B"] * 20,
    "value": (
        # Group A values (tight range)
        [48, 49, 50, 51, 52, 49, 50, 51, 52, 50,
         48, 49, 50, 51, 52, 49, 50, 51, 52, 50]
        +
        # Group B values (tight range, shifted)
        [68, 69, 70, 71, 72, 69, 70, 71, 72, 70,
         68, 69, 70, 71, 72, 69, 70, 71, 72, 70]
    )
}

df_clean = pd.DataFrame(data)

sns.violinplot(data=df_clean, x="group", y="value")
plt.show()
```
