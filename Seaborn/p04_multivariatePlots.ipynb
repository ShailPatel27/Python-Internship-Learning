```python
import seaborn as sns
import matplotlib.pyplot as plt̥̥

df = sns.load_dataset("tips")
df.head()
```

### pairplot — RELATIONSHIPS EVERYWHERE

"I have many numeric columns.
How do all of them relate to each other?"

Instead of making 10 different scatterplots, pairplot does it automatically.

- Each row & column = one numeric feature
- Off-diagonal → scatterplots
- Diagonal → distributions (histograms / KDE)

Diagonal:
- Distribution of a single variable
- Example: frequency of total_bill values

Off-diagonal:
- Relationship between two variables
- Example: total_bill vs tip

```python
sns.pairplot(df)
```

#### Restrict to useful columns

```python
sns.pairplot(df[["total_bill", "tip"]])
```

#### Add grouping with hue

```python
sns.pairplot(
    df[["total_bill", "tip", "size", "sex"]],
    hue="sex"
)
```

### jointplot — FOCUS ON ONE RELATIONSHIP

"I want to deeply understand the relationship between TWO variables."

It shows:
- Relationship plot (center)
- Distribution of each variable (sides)

- Center → scatterplot
- Top → distribution of total_bill
- Right → distribution of tip

```python
sns.jointplot(data=df, x="total_bill", y="tip")
```

#### Different "kind" values

| kind      | Meaning              |
| --------- | -------------------- |
| `scatter` | Raw points (default) |
| `kde`     | Density relationship |
| `hex`     | Closer data = darker hex |
| `reg`     | Regression line      |


```python
sns.jointplot(data=df, x="total_bill", y="tip", kind="kde", fill=True)
# fill only works for kind="kde"
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", kind="hex")
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", kind="reg")
```
