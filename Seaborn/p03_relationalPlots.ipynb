```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.head()
```

### scatterplot — BASIC RELATIONSHIP

- Each dot = one customer
- X-axis → bill amount
- Y-axis → tip
- Upward trend = higher bill → higher tip

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

hue

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.show()
```

Size:
- Bigger dot → larger group
- Smaller dot → fewer people

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    size="size"
)
plt.show()
```

style: different marker styles for different categories

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    style="time"
)
plt.show()
```

### lineplot — TRENDS OVER X

- line represents mean data
- Confidence Interval (CI) represents all data

```python
sns.lineplot(data=df, x="size", y="tip")
plt.show()
```

```python
sns.lineplot(
    data=df,
    x="size",
    y="tip",
    hue="sex",
    errorbar=None
)
plt.show()
```

### relplot — RELATIONSHIP MANAGER

relplot is a figure-level function that manages scatterplots and lineplots.

- kind="scatter" → choose scatterplot
- kind="line" → choose lineplot

```python
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    col="time",
    kind="scatter"
)
```
