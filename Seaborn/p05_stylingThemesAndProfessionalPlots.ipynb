```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips")
df.head()
```

Seaborn controls 3 things:

1. Style → background & grid
2. Context → size (labels, titles, lines)
3. Palette → colors

### set_theme() — MASTER SWITCH

Sets style + context + palette together.

| Style       | Meaning                                        |
| ----------- | ---------------------------------------------- |
| `darkgrid`  | Grey background + grid (default, best for EDA) |
| `whitegrid` | White background + grid                        |
| `white`     | Plain white                                    |
| `dark`      | Dark background                                |
| `ticks`     | Minimal axes                                   |


```python
sns.set_theme(style="darkgrid")
sns.lineplot(data=df, x="tip", y="total_bill")
plt.show()
```

### set_context() — SIZE CONTROL

| Context    | Use case          |
| ---------- | ----------------- |
| `paper`    | Reports           |
| `notebook` | Jupyter (default) |
| `talk`     | Presentations     |
| `poster`   | Big screens       |


```python
sns.lineplot(data=df, x="tip", y="total_bill")
sns.set_context("paper")
plt.show()
```

### Color Palettes — COLORS THAT MAKE SENSE

```python
sns.color_palette() # Shows the current color pallete (colors currently being used)
```

| Palette    | Use                        |
| ---------- | -------------------------- |
| `deep`     | Default                    |
| `muted`    | Soft colors (professional) |
| `pastel`   | Light colors               |
| `viridis`  | Numeric gradients          |
| `coolwarm` | High vs low contrast       |


```python
sns.set_palette("viridis")

sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
sns.color_palette()
```

### Titles, Labels & Figure Size (Matplotlib side)

Seaborn handles plots, but labels come from matplotlib.

```python
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

#### Control figure size

figsize=(width, height)

Units are in inches.

```python
plt.figure(figsize=(2, 5))
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

### Everything Altogether

```python
sns.set_theme(style="whitegrid")
sns.set_context("notebook")
sns.set_palette("muted")

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex"
)

plt.title("Total Bill vs Tip by Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```
