```python
df = sns.load_dataset("tips")
df
```

### Histplot - Frequency Distribution

Shows frequency of values

```python
sns.histplot(df["total_bill"], bins=20, kde=True)
# sns.histplot(data=df, x='total_bill')

# KDE = Kernel Density Estimation
# KDE draws a smooth curve that shows where data values are concentrated.
```

### kdeplot – Smooth Distribution Curve

```python
# the same curve as the one before in histogram with kde attribute
sns.kdeplot(data=df, x="total_bill", hue="sex", fill=True)
plt.show()

# fill=True for filling the color
# fill=False to just print a line
# hue will divide the dataset into different parts and perform the plotting on each part 
```

### displot – Figure-Level Distribution

distplot plots on figure so you can have 2 different plots on the same figure

```python
# Different plots
sns.displot(data=df, x="total_bill", col="sex")
```

histplot with hue:

```python
sns.histplot(data=df, x="total_bill", hue="sex", kde=True)
plt.show()
```
