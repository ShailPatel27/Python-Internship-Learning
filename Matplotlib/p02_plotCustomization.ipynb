```python
import matplotlib.pyplot as plt
```

### Line Styles, Colors & Markers

For colors, you can use:

<li>Names → "red", "blue", "green"</li>
<li>Hex → "#ff5733"</li>
<li>RGB tuple → (0.2, 0.4, 0.6)</li>

Linestyle values: 
<li>'-', 'solid'</li>
<li>'--', 'dashed'</li>
<li>'-.', 'dashdot'</li>
<li>':', 'dotted</li>
<li>'None', ' ', ''</li>

Markers:
| Marker | Shape  |
| ------ | ------ |
| `'o'`  | Circle |
| `'s'`  | Square |
| `'*'`  | Star   |
| `'+'`  | Plus   |
| `'x'`  | X mark |
| `'^'`  | Triangle|


```python
plt.plot([1,2,3], [5,7,4],
         color="red",
         linestyle="--",
         marker="x")
plt.show()
```

### Changing Figure Size

```python
plt.figure(figsize=(2, 1))   # width=2, height=1 inches
plt.plot([1,2,3], [4,5,6])
plt.show()
```

### Setting Axis Limits

```python
plt.xlim(0, 10)     # x-axis values start & end
plt.ylim(0, 100)    # y-axis values start & end
```

### Adding Grid

```python
plt.grid()
```

```python
plt.grid(color="blue", linestyle='-.', linewidth=1)
```

### Transparency

```python
plt.plot([1,2,3], [3,2,9], alpha=0.2)
# alpha=0.0 is invisible → 1.0 is fully visible.
```

### Saving Your Plot

<pre>"plot.pdf" → save as PDF
"plot.jpg" → JPEG

dpi = Dots Per Inch
More dpi = clearer image
Less dpi = blurrier image</pre>
| dpi      | Quality    | Use Case                         |
| -------- | ---------- | -------------------------------- |
| **72**   | Low        | Basic web images, small previews |
| **100**  | Medium     | Normal usage                     |
| **200**  | High       | Good quality for screenshots     |
| **300**  | Very High  | Printing, professional reports   |
| **600+** | Ultra High | Publishing-grade graphics        |


```python
plt.plot([1,2,4],[4,2,3])
plt.savefig("plot.png", dpi=300)
```
