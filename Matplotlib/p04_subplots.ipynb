```python
import matplotlib.pyplot as plt
```

Subplots let you display multiple plots in the same window.

There are two ways to create subplots:
<li>Simple way → plt.subplot()</li>
<li>Better professional way → plt.subplots() (Object-Oriented)</li>

### 1. Using plt.subplot()

plt.subplot(rows, columns, index)

<li>rows → number of rows</li>
<li>columns → number of columns</li>
<li>index → position (1-based index)</li>
(like a table or grid layout)

```python
plt.subplot(2, 1, 1)
plt.plot([1,2,3], [4,5,6])
plt.title("Plot 1")

plt.subplot(2, 1, 2)
plt.plot([1,2,3], [6,5,4])
plt.title("Plot 2")

plt.tight_layout()     #This adjusts spacing so subplots don’t overlap.
plt.show()
```

### 2. Using plt.subplots() (BEST Way)

fig, ax = plt.subplots()

<li>fig = the entire figure</li>
<li>ax = list/array of Axes (actual plots)</li>

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot([1,2,3], [4,5,6])
axes[0].set_title("Left Plot")

axes[1].plot([1,2,3], [6,5,4])
axes[1].set_title("Right Plot")

plt.tight_layout()
plt.show()
```

#### Sharing Axes

```python
# Both charts share the same x-axis labels and scale.
# The top chart doesnt have an x axis
fig, ax = plt.subplots(2, 1, sharex=True)
```

```python
# The right chart doesnt have a y axis
fig, ax = plt.subplots(1, 2, sharey=True)
```

#### Changing Figure Size in Subplots

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 4))
```
