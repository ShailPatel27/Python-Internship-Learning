```python
import matplotlib.pyplot as plt
```

### Using Built-in Themes — plt.style.use()

Matplotlib has many ready-made styles:

- "ggplot"
- "seaborn"
- "dark_background"
- "fast"
- "classic"
- "bmh"
- "fivethirtyeight"

```python
plt.style.use("dark_background")
plt.plot([1, 2, 3, 4], [5,2,3,5])
plt.show()
```

### Color Maps (cmap)

Used mainly in:

- scatter plots
- heatmaps
- images

Popular colormaps:

- "viridis" (default, very clean)
- "plasma"
- "inferno"
- "cool"
- "hot"
- "rainbow" (not recommended for data accuracy)

```python
plt.style.use("default")
plt.grid()

x = [12,5,7,1,7]
y = [1,2,3,4,5]

plt.scatter(x, y, c=y, cmap="viridis")
plt.show()
```

### Adding Text & Annotations

Useful attributes:

- arrowstyle="->"
- fontsize=12
- color="blue"
- ha="right"    // Horizontal Alignment
- va="top"      // Vertical Alignment

```python
x = [1, 5, 6, 2]
y = [1, 2, 3, 4]
fig, ax = plt.subplots()

ax.plot(x, y)
ax.text(2, 3, "(2, 3)", fontsize=15, color="red")
```

### Rotating Ticks

```python
# Used if labels get crowded
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.show()
```

### Adjusting Tick Marks

Change tick positions:

```python
x = [0, 1, 5, 6]
y = [15, 22, 28, 33]

fig, ax = plt.subplots()

ax.plot(x, y, marker="o")

ax.set_xticks([0, 2, 4, 6])
ax.set_yticks([10, 20, 30, 40])
plt.show()
```

Change tick labels:

```python
x = [0, 2, 4, 6]
y = [10, 20, 15, 25]

fig, ax = plt.subplots()

ax.plot(x, y, marker="o")

ax.set_xticks([0, 2, 4, 6])
ax.set_xticklabels(["A", "B", "C", "D"])

plt.show()
```

### Line Width, Marker Size, Alpha (Transparency)

```python
x = [1, 6, 2, 8]
y = [1, 2, 3, 4]

fig, ax = plt.subplots()

ax.plot(x, y, 
        marker="o",
        linewidth=3,
        markersize=15,
        alpha=0.5)
plt.show()
```

### Adding Shapes (Arrows, Rectangles, etc.)

patches.Rectangle((x, y), width, height)

| Part     | Meaning                         |
| -------- | ------------------------------- |
| `(2, 2)` | Bottom-left corner of rectangle |
| `3`      | Width (x-direction)             |
| `4`      | Height (y-direction)            |


```python
import matplotlib.patches as patches

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

fig, ax = plt.subplots()
ax.plot(x, y, marker="o")

# highlight region
rect = patches.Rectangle((1.5, 5.5), 1, 1,
                          edgecolor='yellow',
                          facecolor='none')
ax.add_patch(rect)

circle = patches.Circle((2.5, 7.5), 0.5, edgecolor="red", facecolor="green")
ax.add_patch(circle)

plt.show()
```

```python
x = [1, 2, 2.3, 2.5, 2.75, 4, 4.2, 4.5, 4.75, 5, 6]
y = [1, 1, 1.25, 1.5, 2, 10, 10.3, 10.5, 10.6, 10.6, 10.6]

fig, ax = plt.subplots()
ax.plot(x, y)

ax.axhline(1, color='gray', linestyle='-.')
ax.axvline(4, color='gray', linestyle='--')
plt.show()
```

### Adding Legends (Custom Position)

Positions:

- "upper left"
- "upper right"
- "lower left"
- "center"
- "best"    //default

```python
x = [1, 6, 2, 8]
y = [1, 2, 3, 4]

fig, ax = plt.subplots()

ax.plot(x, y, label="line")
ax.legend(loc="upper right")
```
