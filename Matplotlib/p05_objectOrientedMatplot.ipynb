```python
import matplotlib.pyplot as plt
```

<pre>Instead of using plt.plot() repeatedly (state-based style),
you work with Figure and Axes objects.

This gives you MORE power & cleaner code.<pre>

### Creating Figure & Axes

```python
fig, ax = plt.subplots()

# fig → Entire figure (the window)
# ax → Single axes object (one plot area)
```

```python
fig, ax = plt.subplots()

x = [1,2,3,4]
y = [5,6,7,8]

ax.plot(x, y)
ax.set_title("My Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
```

### Multiple Axes (Multiple Plots)

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot([1,2,3], [3,4,1])
axes[0].set_title("Left Plot")

axes[1].plot([1,2,3], [5,4,8])
axes[1].set_title("Right Plot")
```

### Styling is easier in OO style

```python
fig, ax = plt.subplots()
ax.plot(x, y, color="red", linestyle="--", marker="o")
ax.grid(True)
ax.set_xlim(0, 10)
ax.set_ylim(0, 50)
plt.show()
```

### Adding Multiple Lines to the SAME Axes

```python
x = [1, 2, 3, 4]
y1 = [5,7,1,3]
y2 = [3,0,1,3] 

fig, ax = plt.subplots()

ax.plot(x, y1, label="Product A")
ax.plot(x, y2, label="Product B")

ax.legend()

plt.show()
```

### Adding Text / Annotation

```python
x = [1,2,3,4]
y = [5,6,7,8]

fig, ax = plt.subplots()

ax.plot(x, y, color="red", linestyle="--", marker="o")
ax.grid(True)

# Highlight actual data point
ax.annotate("Peak Value",
            xy=(4, 8),  #pointed position
            xytext=(5, 15), #text position
            arrowprops=dict(arrowstyle="->"))

ax.set_xlim(0, 10)
ax.set_ylim(0, 20)

plt.show()
```
