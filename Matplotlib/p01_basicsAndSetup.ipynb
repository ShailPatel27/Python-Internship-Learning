Matplotlib is Python’s main plotting library. <br>
It lets you create all types of charts—line, bar, scatter, histograms, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 15, 25])
plt.show()

# First list = X-axis
# Second list = Y-axis
# plt.show() displays the plot window

# so, it is ({1,10}, {2,20}, {3,15}, {4,25})
```

```python
plt.plot([1,2,3],[5,7,4], label="Sales")    #label name is used in legend
plt.title("Monthly Sales")  #title of the plot
plt.xlabel("Month")         #x axis label
plt.ylabel("Units")         #y axis label
plt.legend()                # shows labels
plt.show()
```

### Figure vs Axes

<h5>Figure</h5>
<li>The entire “canvas” (overall window)</li>

```python
fig = plt.figure()
```

<h5>Axes</h5>

<li>The actual chart area where the data is drawn</li>
<li>A figure can have MANY axes (subplots)</li>

```python
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_subplot(10,20,30)
```
