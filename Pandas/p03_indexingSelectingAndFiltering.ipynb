### Filtering Rows

```python
df1[df1["Age"] > 25]
```

```python
df1[df1["City"] == "Mumbai"].loc[:, ["Name","City"]]
```

```python
df[df["First Name"].str.contains("Eli")]
```

```python
df1[(df1["Age"] > 25) & (df1["City"] == "Ahmedabad")]
```

```python
df1[(df1["City"] == "Delhi") | (df1["City"] == "Mumbai")]
```

### Resetting Index

```python
# After filtering, index may look weird (like 0, 3, 7).
df1 = df1[(df1["City"] == "Delhi") | (df1["City"] == "Ahmedabad")]
df1
```

```python
# We fix this using:
df1.reset_index(drop=True, inplace=True)
df1
```
