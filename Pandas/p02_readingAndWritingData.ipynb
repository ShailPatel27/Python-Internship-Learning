```python
import pandas as pd
```

### CSV

#### Reading Data

```python
df = pd.read_csv("data/people.csv")
```

```python
df.head()
```

##### Optional Parameters

```python
nm = ["SrNo", "Fname", "Lname", "Gender", "Email", "PhNo", "DOB", "Job"]

df1 = pd.read_csv("data/people.csv", nrows=7, header=0, names=nm)    #No column names

# nrows: reads only the first n rows
# header: defines which row is the header. If there are no headers in your csv, the first row will be taken which is wrong. Instead write header=None so pandas will automatically write header values 0->n.
# names: changes names of header values (columns). If another value is giver in header, eg. 3, all rows above 3 are not considered and the 4th row is replaced with names defined.

df1
```

#### Writing Data

```python
df.to_csv("new data/data.csv", index=False)
# index=False is written because pandas auto indexes rows, which we dont want since we already have a column for index
```
