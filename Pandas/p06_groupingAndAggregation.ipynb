```python
import pandas as pd
```

```python
df = pd.read_csv("data/people.csv")
df.head(2)
# df.drop(["Unnamed: 0","Unnamed: 0.1","Unnamed: 0.2"], axis=1, inplace=True)
# df.to_csv("data/people.csv", index=None)
display(df)
```

### Basic GroupBy

Group by creates groups based on the value given. <br>
Then you can perform operations on those values.

```python
display(df.groupby("City").size())
# How many rows per city
```

```python
df.groupby("City")["Age"].mean()
# Average per city
```

### Group by multiple columns

```python
df.groupby(["City", "Sex"])["Age"].mean()
```

```python
df.groupby("City")["Salary"].sum()
```

### Multiple aggregations at once

```python
df.groupby("City")["Salary"].agg(["mean", "sum", "max", "min"])
```

### Aggregating many columns at once

```python
df.groupby("City").agg({
    "Age": "mean",
    "Salary": ["min", "max"],
})
```

### Reset index

```python
# Creates an index for group by output
df.groupby("City")["Age"].mean().reset_index()
```

### Sorting group results

```python
df.groupby("City")["Age"].mean().sort_values()
# Sorts by values in ascending order
```

```python
df.groupby("City")["Age"].mean().sort_values(ascending=False)
```

### Filter groups

```python
df.groupby("Salary").filter(lambda x: len(x) > 6)
# Rows belonging to Salary values that appear more than 6 times.
```
