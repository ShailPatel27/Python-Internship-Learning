### CONCATENATION (stacking DataFrames)

Used when you want to:
<ul>
    <li>Add more rows to a DataFrame</li>
    <li>Add more columns to a DataFrame</li>
</ul>

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "City": ["Delhi", "Mumbai", "Chennai", "Kolkata"]
})
display(df1)
```

```python
df2 = pd.DataFrame({
    "ID": [5, 6, 7],
    "Name": ["Eve", "Frank", "Grace"],
    "City": ["Mumbai", "Delhi", "Pune"]
})
display(df2)
```

```python
# Vertical concat (add rows)
pd.concat([df1, df2], axis=0)
```

```python
# Horizontal concat (add columns)
pd.concat([df1, df2], axis=1)
```

### MERGE (SQL-style joins)

```python
df2_mod = pd.DataFrame({
    "ID": [2, 4, 7],
    "Department": ["HR", "Finance", "IT"]
})
df2_mod
```

##### INNER JOIN

```python
inner_join = pd.merge(df1, df2_mod, on="ID", how="inner")   #takes common of both
inner_join
```

##### LEFT JOIN

```python
left_join = pd.merge(df1, df2_mod, on="ID", how="left") #takes all the values of the left one (df1)
# left_join = pd.merge(df2_mod, df1, on="ID", how="left") #takes all the values of the left one (df2_mod)
left_join
```

##### RIGHT JOIN

```python
right_join = pd.merge(df1, df2_mod, on="ID", how="right")   #takes all the values of the right one(df2_mod)
right_join
```

##### FULL OUTER JOIN

```python
outer_join = pd.merge(df1, df2_mod, on="ID", how="outer")   #takes all values
outer_join
```

### JOIN (index-based merge)

```python
df1_idx = df1.set_index("ID")
df2_idx = df2_mod.set_index("ID")

df1_idx.join(df2_idx, how="left")
```
