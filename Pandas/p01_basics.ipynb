### Series
Series is a 1D Labelled Array

#### Initializing

```python
# auto index
s = pd.Series([10,20,30,40])
print(s)
```

```python
# custom index
s2 = pd.Series([10,20,30,40], [7,1,4,2])
print(s, end="\n\n")

s3 = pd.Series([10,20,30,40], ['a','b','c','d'])
print(s)
```

#### Accessing Values

```python
# You can access values through index
print(s3['a'])
print(s[1])
```

#### Vector Like Operations

```python
print(s+5)  # Add 5 to all values of 5
print(s*2)  # Multiply all values of s with 2
print(s.mean()) # Avg of all values of s
```

### Dataframe
Dataframe is a collection of series. <br>
It is a 2D table of rows and columns.

```python
data = {
    "Name": ["Shail", "Raj", "Rohan"],
    "Age": [21, 18, 32],
    "City": ["Ahmedabad", "Mumbai", "Rajkot"]
}

df = pd.DataFrame(data)
df
```

#### Basic DataFrame Functions

```python
df.head()   #Prints first 5 rows
```

```python
df.tail(2)  #Prints Last 2 rows
```
