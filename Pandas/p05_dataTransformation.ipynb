```python
import pandas as pd
import numpy as np
```

### Sorting Data

```python
df = pd.read_csv("data/people.csv")
```

```python
display(df.head())
```

```python
df.sort_values("First Name")
```

```python
df.sort_values("First Name", ascending=False)
```

```python
df.sort_values(["First Name", "Last Name"])
```

### Adding Columns

```python
df["Full Name"] = df["First Name"] + " " + df["Last Name"]
display(df)
```

```python
df["DoubleAge"] = df["Age"] * 2
display(df.head(3))
```

```python
df["Adult"] = df["Age"].apply(lambda x: True if x>=21 else False)
display(df[["Age", "Adult"]].sort_values("Adult"))
```

### Removing Columns

```python
df.drop("DoubleAge", axis=1, inplace=True)
# axis=0 -> rows
# axis=1 -> cols
```

```python
df.drop(["Job Title", "Phone"], axis=1)
```

### apply() + lambda

```python
df["Age"].apply(lambda x: x * 5)
# display(df["Age"].sort_values(ascending=True))
```

```python
df["Full Name"] = df["Full Name"].apply(lambda x: x.strip().upper())
# This will remove any extra spaces in the beginning and the end, not in between.
```

```python
df["Retirement"] = df["Age"].apply(lambda x: "Yes" if x > 55 else "No")
display(df[["Age", "Retirement"]].sort_values("Retirement",ascending=True))
```

### map() — replace values using a dictionary

```python
df["Sex"] = df["Sex"].map({"Male": "", "Female": "F"})
display(df["Sex"])
```

```python
# Replacing Unknown Values
df["Sex"] = df["Sex"].map({"": "Unknown"}).fillna(df["Sex"])
display(df["Sex"])
```

### Binning / Categorizing numeric values

```python
# Turn a numeric column into categories.
df["Age Group"] = pd.cut(
    df["Age"],          #The column we are categorizing
    bins=[0, 18, 50, 100],  #Age groups
    labels=["Child", "Adult", "Senior"] #Categories
)

# 0-18: Child
# 19-50: Adult
# 51-100: Senior

display(df[["Age", "Age Group"]].head(10))
```

### Renaming Columns
