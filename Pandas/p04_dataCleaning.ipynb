### Checking Missing Values (NaN)

```python
# na stands for Not Available

df.isna()
# Checks if any value is missing.
# If missing, it returns True in that cell
```

```python
df.isna().sum()
# Count missing values per column
```

### dropna() — Remove rows with NaN

```python
df.dropna()
# Will not change values. You will have to take it in a variable or reassign it.
```

```python
df.dropna(how="all")
# Remove rows only if all columns are NaN
```

```python
# df1 = df    # Will create a pointer ❌
df1 = df.copy() # Will create a new copy ✅
df1.dropna(inplace=True)
# Changes the original DataFrame

df1
```

### fillna() — Replace missing values

```python
# Fill NaN with a fixed value
df3 = df.fillna(0)
df3.isna().sum()
```

```python
df3 = df.copy()

# Convert all non numeric data(outliers/errors) to numeric
df3["Quantity"] = pd.to_numeric(df3["Quantity"], errors="coerce")

# Fill a column with mean value
df3["Quantity"] = df3["Quantity"].fillna(df3["Quantity"].mean())
df3.isna().sum()
```

### replace() — Replace ANY value

```python
df2["Gender"] = df2["Gender"].replace({"Male": "M", "Female": "F"})
df2["Gender"]
```

```python
# Replace multiple values
df2 = df2.replace(["0", "-1", "1"], value="Error")
df2[df2["Age"].str.contains("Error")]
```
